```python
bank_data ['Sex'] = bank_data ['Sex'].fillna(bank_data['Sex'].mode()[0])
bank_data ['MaritalStatus'] = bank_data ['MaritalStatus'].fillna(bank_data['MaritalStatus'].mode()[0])
bank_data ['Province'] = bank_data ['Province'].fillna(bank_data['Province'].mode()[0])
bank_data ['Age'] = bank_data ['Age'].fillna(bank_data['Age'].mode()[0])


bank_data ['A_3month'] = bank_data ['A_3month'].fillna(bank_data['A_3month'].mean())
bank_data ['A_max_3month'] = bank_data ['A_max_3month'].fillna(bank_data['A_max_3month'].mean())
bank_data ['A_min_3month'] = bank_data ['A_3month'].fillna(bank_data['A_3month'].mean())
```

<p><b><i>Removing Outliers<i><p>

```python
def threhold(z,d):
    for i in np.arange(1,5,0.2):
        data = d.copy()
        data = data[(z < i).all(axis = 1)]
        loss = (d.shape[0] - data.shape[0]) / d.shape[0]*100
        print('With threshold {} data loss is {}%'.format(np.round(i,1),np.round(loss,2))) 

z = np.abs(zscore(bank_data))
threhold(z,bank_data)
```

```python
bank_data = bank_data[(z < 2.0).all(axis=1)]
```

```python
x_train,x_not_train,y_train,y_not_train = train_test_split(x,y,test_size = 0.3, random_state =42,stratify=y)
print("Train dataset shape: {0}, \nRemaining dataset shape: {1}".format(x_train.shape, x_not_train.shape))
```

<p><i><b>To evaluate performances of all the models</b></p>

```python
def performance(p,y_test,m,x_test,s):
    print('-----------',m,'-----------')
    print('Accuracy',np.round(accuracy_score(p,y_test),4))
    print('Mean of Cross Validation Score',np.round(s.mean(),4))
    print('AUC_ROC Score',np.round(roc_auc_score(y_test,m.predict_proba(x_test)[:,1]),4))
    print('Confusion Matrix')
    print(confusion_matrix(p,y_test))
    print('Classification Report')
    print(classification_report(p,y_test))
```

<p><i><b>Creating a list of models which will be created one by one</b></p>

```python
models=[LogisticRegression()
        ,DecisionTreeClassifier(),
        RandomForestClassifier()]
```

<p><i><b>Creates and trains model from the models list</b></p>

```python
def createmodel(trainx,valx,trainy,valy):
    for i in models:
        model = i
        model.fit(trainx,trainy)
        p = model.predict(valx)
        score = cross_val_score(model,x,y,cv = 10)
        performance(p, valy, model, valx, score) 
```

```python
createmodel(x_train, x_val, y_train, y_val)
```
